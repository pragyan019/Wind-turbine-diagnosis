```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
model = RandomForestRegressor()
X_train, X_test, y_train, y_test = train_test_split(df.drop('Active.Power' ,axis=1), df['Active.Power'], train_size=.8)
model.fit(X_train, y_train)
```

```python
preds = pd.DataFrame()
preds['preds'] = model.predict(X_test)
preds['actual'] = y_test.values
preds['MAPE'] = ((preds['actual'] - preds['preds'])/preds['actual']).abs()
```

```python
preds['decile'] = pd.qcut(preds['MAPE'],10, labels = range(1,11))
preds.groupby('decile').agg({'MAPE':'mean'})
```

```python
df_feats = pd.DataFrame()
df_feats['feats'] = df.drop('Hydraulic.Oil.Temp',axis=1).columns
df_feats['importance'] = model.feature_importances_
df_feats.sort_values('importance', ascending=False)
```
